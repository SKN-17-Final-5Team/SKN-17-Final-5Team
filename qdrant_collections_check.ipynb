# Qdrant 컬렉션 체크

```python
import os
from dotenv import load_dotenv
from qdrant_client import QdrantClient

load_dotenv()

client = QdrantClient(
    url=os.getenv("QDRANT_URL"),
    api_key=os.getenv("QDRANT_API_KEY"),
    timeout=60
)

collections_resp = client.get_collections()

total_cnt = 0
for coll in collections_resp.collections:
    total_cnt += 1
    info = client.get_collection(coll.name)
    vectors = info.vectors_count or {}
    total_vectors = sum(v.count for v in vectors.values()) if vectors else "n/a"
    
    print(f"- {coll.name} | points: {info.points_count} | vectors: {total_vectors}")

print(f"\nTotal # of collections: {total_cnt}")
```
